<h1>Our first Plot (bar)</h1>
<li>Import the necessary libraries</li>
<li>Get the data</li>
<li>Copy the data into a dataframe</li>
<li>Set the index</li>
<li>Create a plot</li>
<li>Show the plot</li>

```python
import pandas as pd
import matplotlib.pyplot as plt
language = ['Java', 'C', 'Python', 'C++', 'C#', 'VB.NET', 'JavaScript', 'PHP', 'SQL', 'Swift']
percentage = [17.358, 16.766, 9.345, 6.164, 5.927, 5.862, 2.060, 2.018, 1.526, 1.460]

df = pd.DataFrame([language,percentage], ["Language", "%"])
df = df.T
df.set_index('Language', inplace=True)
df.head()

bars = plt.bar(df.index, df['%'])
plt.show()
```

<h1>Create Some Patterns</h1>
<li>Create a list of patterns</li>
<li>Apply the patterns to each bar</li>

```python
fig = plt.figure(figsize=(10,6))
bars = plt.bar(df.index, df['%'])
hatches = ['+', '\\', '|', '*', '**', '.', 'O', '/', '-', '//']

for i in range(10):
    bars[i].set_hatch(hatches[i])
plt.show()
```

<h1>Apply Some Colours</h1>
<li>Create a list of colours</li>
<li>Apply the colours to each bar</li>

```python
fig = plt.figure(figsize=(10,6))

bars = plt.bar(df.index, df['%'])
colours = ['red', 'yellow', 'pink', 'orange', 'green', 'purple', 'blue', 'indigo', 'black', 'white']

for i in range(10):
    bars[i].set_color(colours[i])
plt.show()
```

<h1>Adding some information to the plot</h1>
<li>A title </li>
<li>x,y labels (title, xlabel, ylabel)</li>

```python
fig = plt.figure(figsize=(10,6))
bars = plt.bar(df.index, df['%'])

plt.title("Popular Programming Languages")
plt.xlabel("Language")
plt.ylabel("Percentage")

colours = ['red', 'yellow', 'pink', 'orange', 'green', 'purple', 'blue', 'indigo', 'black', 'white']

for i in range(10):
    bars[i].set_color(colours[i])
plt.show()
```

<h1>Saving a Plot (graph)</h1>

```python
fig = plt.figure(figsize=(10,6))
plt.title("Popular Programming Languages")
plt.xlabel("Language")
plt.ylabel("Percentage")
bars = plt.bar(df.index, df['%'])
colours = ['red', 'yellow', 'pink', 'orange', 'green', 'purple', 'blue', 'indigo', 'black', 'white']

for i in range(10):
    bars[i].set_color(colours[i])


plt.savefig('myPlot2.png')

plt.show()
```

<h1>Creating a Horizontal Bar Plot </h1>

```python
fig = plt.figure(figsize=(10,6))
bars = plt.barh(df.index, df['%'])
plt.title("Popular Programming Languages")
plt.ylabel("Language")
plt.xlabel("Percentage")
colours = ['red', 'yellow', 'pink', 'orange', 'green', 'purple', 'blue', 'indigo', 'black', 'white']

for i in range(10):
    bars[i].set_color(colours[i])


plt.savefig('myPlot2.png')

plt.show()
```
