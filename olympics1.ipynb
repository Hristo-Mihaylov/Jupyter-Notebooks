<h1>Reading a Datafile into a DataFrame</h1>
The basic process of loading data from a CSV file into a Pandas DataFrame is achieved using the read_csv method.  We specify the pandas alias and the read_csv method (seperated with a dot).  The basic parameter for reading a file is to specify the path and filename.

```python
df = pd.read_csv('olympics.csv')
```

```python
df.head(10)
```

<h1>Setting the Row Index using column number</h1>
We can use a column index to specify which column should be the row index.  In this case we want column 0 to become the row index (the column with the Contries).  We can also used labels for the column index.

```python
df = pd.read_csv('olympics.csv', index_col=0)
df.head(10)
```

<h1>Skip Rows</h1>
The skip rows attribute will allow us to skip rows from the file that is being read into the dataframe.  This is useful if the datafile contains metadata.  It is also useful for setting the column names to the column names in the datafile.

```python
df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
df.head(10)
```

The column names are stored in an attribute named columns which belongs to the DataFrame object.  To view the column names we can specify the DataFrame object name, a dot and then the column attribute.

```python
df.columns
```

<h1>Renaming Column Names</h1>
The rename method will allow us to rename a column name.  We specify the original column name, a colon, then the new column name.  The columns are stored in an attribute named columns, thus we use the assignment operator to assign the new value to columns.

In this example we are creating and empty string for the new column name.  Then we are checking if the first two characters of the value in columns (which we copy into col for each column value) is equal to '01', it is is we append 'G' to the empty string, the elif catches if the column start of the column is '02' or '03' and if it is it will be replaced with 'S' or 'B'.

The next if checks the end of the column name.  If the final value is equal to !, then 'S' is appended to the the value already in column_name.  If the last three values of column name are '!.1' or '!.2' then we append 'W' or 'T'.

The final if checkes if the column_name has been updated (no longer an empty string).  If it has been updated the original column value is replaced by the new column value.  

The inplace attribute saves the changes to the column attribute of the dataframe in place (at this time).

```python
for col in df.columns:
    column_name = ""
    if col[0:2] == '01':
        column_name += 'G'
    elif col[0:2] == '02':
        column_name += 'S'
    elif col[0:2] == '03':
        column_name += 'B'
        
    if col[-1:] == '!':
        column_name += 'S'
    elif col[-3:] == '!.1':
        column_name += 'W'
    elif col[-3:] == '!.2':
        column_name += 'T'
    
    if column_name != "":
        df.rename(columns={col:column_name}, inplace=True)
```

<h1>Boolean Mask</h1>
There are several ways to create a Boolean mask, however, you will almost always use the comparison operators on just a single columnor Series of data. In this example we are interested in countries achieving a gold in the summer olympics, so we use the GS column of the data frame and the evaluation greater than zero.  False means a gold medal was not achieved, True means at least one gold medal was achieved.

```python
df['GS'] > 0
```

<h1>Using a Boolean Mask to filter data into a new DataFrame</h1>
Using the where method, we can apply a Boolean mask to a dataframe.  In this example we copy the results into a new DataFrame named df_gold_only.

The head method is then used to examine the first three rows of the data frame.  All countries are copied into the new DataFrame, however, those for those countries that evaluated to false, the original data is replaced by NaN (not a number).

```python
df_gold_only = df.where(df['GS'] > 0)
df_gold_only.head(3)
```

<h1>Dropping Missing Values</h1>
We can remove missing values from a DataFrame by using the dropna method.  By specifying a value for the axis attribute we can determine if rows or columns which contain missing values are removed.  Care should be taken when using dropna because it will drop the row or column if only one value is missing.  Sometimes you might want to replace a missing value with another value rather than drop.

```python
df_gold_only = df_gold_only.dropna()
df_gold_only.head(3)
```

<h1>Using the Index Operator to Apply a Filter</h1>
Another approach to filtering data based on Boolean logic is to use the index operator.  In this example we use the index to only extract data from the df DataFrame for those rows with a value greater than 0 in the SS column.

```python
df_silver_only = df[df['SS'] > 0]
df_silver_only.head(3)
```

Multiple Boolean logic statements can be applied by using and/or operators.  In this example we are creating a new DataFrame (actually we are overwriting the df_gold_only DataFrame) that filters out all the rows that have a value greater than 0 for either column GS or GW (Gold Summer, Gold Winter).  We could have used GT (Gold Total), but this make the example of using the OR operator redundant. 

```python
df_gold_only = df[(df['GS'] > 0) | (df['GW'] > 0)]
df_gold_only.head(60)
```

<h1>KeyError and loc</h1>
When you know a key is in your dataset (because you can see it), it can be quite bewildering to recieve key errors.  In this example, to print the row information for ‘Liechtenstein (LIE)’ we should just be able to:

df_gold_only.loc['Liechtenstein (LIE)']

However, in this case we have different representations for space.  If you output an index using the index method and provide a numeric index, ie.

df_gold_only.index[5] you would expect the output to be:  Austria (AUT), since that is what you can see in the DataFrame.  However, it actually outputs 'Austria\xa0(AUT)', '\xa0' is a non-breaking space in Latin1 (ISO 8859-1), which is not the same as ' '.  Therefore, using the loc method results in a key error.

To resolve this, we can just replace all the values '\xa0' with a space.

Then, our loc method will work as expected.

```python
df_gold_only.index[5]
```

```python
for value in df_gold_only.index:
    tmp = value.replace('\xa0', ' ')
    df_gold_only = df_gold_only.rename(index={value:tmp})
```

```python
df_gold_only.loc['Liechtenstein (LIE)']
```
