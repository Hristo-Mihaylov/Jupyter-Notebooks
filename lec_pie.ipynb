<h1>Import the necessary libraries </h1>
<li>Get the data</li>
<li>Copy the data into a dataframe</li>
<li>Set the index</li>

```python
import pandas as pd
import matplotlib.pyplot as plt
language = ['Java', 'C', 'Python', 'C++', 'C#', 'VB.NET', 'JavaScript', 'PHP', 'SQL', 'Swift']
percentage = [17.358, 16.766, 9.345, 6.164, 5.927, 5.862, 2.060, 2.018, 1.526, 1.460]

df = pd.DataFrame([language,percentage], ["Language", "%"])
df = df.T
df.set_index('Language', inplace=True)
df.head()
```

<h1>Do the Plot</h1>
<li>Create a plot</li>
<li>Show the plot</li>
<li>Add the explode feature</li>
<li>Add percentages to the slices</li>
<li>Add some labels</li>

```python
fig = plt.figure(figsize=(6,20))
explode = [0.05]*10
plt.pie(df['%'], labels=df.index, autopct='%.0f %%', explode=explode)
plt.title("Popular programming languages")
plt.show()
```
