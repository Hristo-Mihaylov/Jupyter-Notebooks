```python
df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
df.head(10)
```

```python
df.columns
```

```python
for col in df.columns:
    column_name = ""
    if col[0:2] == '01':
        column_name += 'G'
    elif col[0:2] == '02':
        column_name += 'S'
    elif col[0:2] == '03':
        column_name += 'B'
        
    if col[-1:] == '!':
        column_name += 'S'
    elif col[-3:] == '!.1':
        column_name += 'W'
    elif col[-3:] == '!.2':
        column_name += 'T'
    
    if column_name != "":
        df.rename(columns={col:column_name}, inplace=True)
```
